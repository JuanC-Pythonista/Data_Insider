```python
import requests
import pandas as pd
```

## **2. Obtenci√≥n de los datos**

```python
#Apple
# Se lee un archivo .txt donde se encuentra el key privado de la API
with open("InfoIgnore/pass.txt",'r') as f:
    pwd= f.read()
# Se lee el API, se genera el requests y se descargan los datos de la API
url1 = "https://api.polygon.io/v2/aggs/ticker/AAPL/range/1/week/2022-01-01/2022-12-31?adjusted=true&sort=asc&apiKey="+pwd
headers1 = {"Accept-Encoding": "gzip, deflate"}
response1 = requests.get(url1, headers=headers1)
Apple = response1.json()
print(response1)
print(Apple)
Apple.keys()
```

```python
DataSetApple= pd.DataFrame(Apple['results'])
DataSetApple.head(10)
```
